# Data Analytics

O objetivo da análise foi utilizar metodos nativos com auxilio do Pandas para uma análise rápida de duas fontes distintas sobre operadora de Streaming Netflix.
Uma fonte é oficial e outra sintética gerada por IA. Desta forma, o material é apenas para prática de estudos.

```python
# Carregar biblioteca para a sua sessão Python
import pandas as pd
import seaborn as sns
```

```python
# Carregar o primeiro arquivo
arquivo = '/content/Netflix Userbase.csv'
df = pd.read_csv(arquivo)
df.head()
```

```python
# dados faltantes
df.isnull().sum().to_frame().T
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
arquivo2 = '/content/netflix_titles.csv'
df2 = pd.read_csv(arquivo2)
df2.head()
```

```python
df2.isnull().sum().to_frame().T
```

```python
df2.fillna('')
```

```python
df2.info()
```

```python
# primeira base de dados / somente Brasil
df_brazil = df[df.Country == "Brazil"]
df_brazil
```

```python
# segunda base de dados / somente Brasil
df2_brazil = df2[df2.country == "Brazil"]

```

```python
df2_brazil.fillna('-')
```

```python
# cores marca
sns.palplot(['#221f1f', '#b20710', '#e50914','#f5f5f1'])


```

```python
# Estatísticas descritivas de dados numéricos
df_brazil.describe().round(1)
```

```python
# Estatísticas das variávels categóricas
df_brazil.describe(include='object')
```

```python
df2_brazil.describe().round(1)
```

```python
df2_brazil.describe(include='object')
```

```python
# Média de idade dos assinantes / Brasil
df_brazil.groupby('Country')['Age'].mean().sort_values()
```

```python
# Total de valores de lancamentos/Ano no Brasil
df2_brazil['release_year'].value_counts()
```

```python
# Total de Titulos / Diretor no Brasil
df2_brazil['director'].value_counts()
```

Insights:



*   **Média Idade** assinantes do brasil: 38 anos.
*   Plano mais assinado: **Basic**.
*   Dispositivo mais acessado: **Smartphone**.
*   **77 títulos** e produções brasileiras.
*   **Stand Up comedy** é o gênero mais assistido.
*   **Cidade de Deus** é o filme brasileiro mais assistido.
*   **Filmes 65% e 35% Programas de TV** (séries, documentários, reality show etc.).
